```python
import pandas as pd
wine = pd.read_csv('wine/wine.data', names=['class','Alcohol', 'Malicacid', 'Ash', 'Alcalinity_of_ash', 'Magnesium'])
print(wine.head())
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

# scatter the Alcohol against the Malicacid
ax.scatter(wine['Alcohol'], wine['Malicacid'])
# set a title and labels
ax.set_title('Wine Dataset')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Malicacid')
```

```python
# create color dictionary
colors = {'Wine-Alchohol':'r', 'Wine-Malicacid':'g', 'Wine-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(wine['Alcohol'])):
    ax.scatter(wine['Alcohol'][i], wine['Malicacid'][i],color=colors[wine['class'][i]])
# set a title and labels
ax.set_title('Wine Dataset')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Malicacid')
```

```python
# get columns to plot
columns = wine.columns.drop(['class'])
# create x data
x_data = range(0, wine.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, wine[column])
# set title and legend
ax.set_title('Wine Dataset')
ax.legend()
```
